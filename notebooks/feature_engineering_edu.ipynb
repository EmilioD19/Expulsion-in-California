```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filepath = "/Users/emiliodulay/Documents/Programming/Student Success Classification/student-success/data/top5_counties.csv"
df = pd.read_csv(filepath)
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.drop(["AcademicYear", "DistrictName", "SchoolName"], axis = 1)
```

```python
df.head()
```

```python
# Need to subset demogrpahic "Total"
df = df[df["Demographic"] != "Total"].reset_index(drop = True)
```

```python
df.head()
```

```python
# one-hot encode demographic
df = pd.get_dummies(df, 
                    columns = ["Demographic"],
                    prefix = "",
                    prefix_sep = "",
                    dtype = int)
```

```python
df.head()
```

```python
# one-hot encode AggregateLevel
df = pd.get_dummies(df,
                    columns = ["AggregateLevel"],
                    prefix = "is",
                    prefix_sep = "_",
                    dtype = int)
```

```python
df.head()
```

```python
df.to_csv("top5_counties_feature.csv", index = False)
```
